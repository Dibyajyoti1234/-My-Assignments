```python
#1
Matplotlib is an exceptional visualization library in Python primarily used for creating 2D plots of arrays.
1. Line plot
2. Bar plot
3. Histogram
4. Scatter Plot
5. 3D Plot
```

```python
#2
# Scatter Plot is a type of plot that displays the relationship between two variables in a dataset. 
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y, c="blue", label="Data Points")
plt.title("Scatter Plot of x vs. y")  
plt.xlabel("x")  
plt.ylabel("y") 
plt.grid(True)  
plt.legend()  
plt.show()
```

```python
#3
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label="Line 1")
plt.title("Line 1")
plt.subplot(2, 2, 2)
plt.plot(x2, y2, label="Line 2")
plt.title("Line 2")
plt.subplot(2, 2, 3)
plt.plot(x3, y3, label="Line 3")
plt.title("Line 3")
plt.subplot(2, 2, 4)
plt.plot(x4, y4, label="Line 4")
plt.title("Line 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.tight_layout()
plt.show()
```

```python
#4
# A bar plot is is a type of graph that visually represents categorical data. It uses rectangular bars, where the height or length of each bar corresponds to the values associated with specific categories. 

# bar plot
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit, color='maroon', width=0.4)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Profit by Company")
plt.show()

# Horizontal plot
import matplotlib.pyplot as plt
import numpy as np
plt.barh(company, profit, color='purple')
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Profit by Company (Horizontal)")
plt.gca().invert_yaxis()  # Invert y-axis for readability
plt.show()
```
