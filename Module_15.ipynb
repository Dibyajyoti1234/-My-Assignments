```python
#1
def flatten_list(lst):
    """
    Flattens a nested list and returns a flat list containing all elements.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten_list(item))
        elif isinstance(item, dict):
            # Extract numeric keys and values from the dictionary
            for key, value in item.items():
                if isinstance(key, (int, float)):
                    flat_list.append(key)
                if isinstance(value, (int, float)):
                    flat_list.append(value)
        else:
            flat_list.append(item)
    return flat_list

def product_of_numbers(lst):
    """
    Calculates the product of all numbers in a flat list.
    """
    flat_lst = flatten_list(lst)
    product = 1
    for num in flat_lst:
        if isinstance(num, (int, float)):
            product *= num
    return product


list1 = [1, 2, 3, 4, [44, 55, 66, True], False, (34, 56, 78, 89, 34),
         {1, 2, 3, 3, 2, 1}, {1: 34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)},
         [56, 'data science'], 'Machine Learning']


result = product_of_numbers(list1)


print(f"The product of all numbers in the list is: {result}")
```

```python
#2
def encrypt_message(input_sentence):
    """
    Encrypts a message by replacing each letter with its corresponding encrypted letter.
    Whitespace is replaced with a dollar sign ('$').
    Punctuation marks remain unchanged.
    """
    # Define the mapping from letters to their encrypted counterparts
    letter_mapping = {
        'a': 'z', 'b': 'y', 'c': 'x', 'd': 'w', 'e': 'v',
        'f': 'u', 'g': 't', 'h': 's', 'i': 'r', 'j': 'q',
        'k': 'p', 'l': 'o', 'm': 'n', 'n': 'm', 'o': 'l',
        'p': 'k', 'q': 'j', 'r': 'i', 's': 'h', 't': 'g',
        'u': 'f', 'v': 'e', 'w': 'd', 'x': 'c', 'y': 'b',
        'z': 'a'
    }

    # Convert input sentence to lowercase
    input_sentence = input_sentence.lower()

    # Initialize an empty encrypted sentence
    encrypted_sentence = ""

    
    for char in input_sentence:
        if char.isalpha():
        
            encrypted_char = letter_mapping.get(char, char)
            encrypted_sentence += encrypted_char
        elif char.isspace():
            
            encrypted_sentence += "$"
        else:
            
            encrypted_sentence += char

    return encrypted_sentence


input_sentence = "I want to become a Data Scientist."


encrypted_result = encrypt_message(input_sentence)


print(f"Encrypted sentence: {encrypted_result}")
```
