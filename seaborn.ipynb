```python
#1
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five plots that can be plotted using Seaborn library and their uses:

Line Plot: A line plot is a graph that displays data as points or markers connected by straight lines. It is used to show trends over time, compare two or more variables, or to visualize a continuous relationship between two variables .

Bar Plot: A bar plot represents an estimate of the central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars. It is used to compare the values of different categories .

Histogram: A histogram is a graph showing frequency distributions. It is used to represent the distribution of continuous data .

Box Plot: A box plot is a method for graphically demonstrating the location, dispersion, and skewness of a dataset. It is used to identify outliers and to compare distributions .

Heat Map: A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors. It is used to visualize complex data sets and to identify patterns in the data 
```

```python
#2
import seaborn as sns

fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
#3
import seaborn as sns
titanic = sns.load_dataset("titanic")
sns.boxplot(x="pclass", y="age", data=titanic)
sns.boxplot(x="pclass", y="fare", data=titanic)
```

```python
#4
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x="price", hue="cut")
```

```python
#5
import seaborn as sns
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
```

```python
#6
import seaborn as sns
flights = sns.load_dataset("flights")
sns.heatmap(flights.pivot("month", "year", "passengers"), cmap="YlGnBu")
```
