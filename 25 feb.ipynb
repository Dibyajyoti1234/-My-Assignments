```python
#2
loc (Label-based Selection):
loc stands for label-based selection.
It selects rows and columns based on labels (i.e., row and column names).
You pass the names of the rows or columns you want to select.
The last element of the range passed to loc is included.
It can accept boolean data for filtering.

iloc (Integer-based Selection):
iloc stands for integer-based selection.
It selects rows and columns based on integer positions (i.e., row and column indices).
You pass the integer indices of the rows or columns you want to select.
The last element of the range passed to iloc is excluded.
It does not accept boolean data for filtering.

```

```python
#8
import pandas as pd

def main():
    try:
        
        date1_str = input("Enter the first date (YYYY-MM-DD): ")
        date2_str = input("Enter the second date (YYYY-MM-DD): ")
        date1 = pd.to_datetime(date1_str)
        date2 = pd.to_datetime(date2_str)
        time_difference = date2 - date1
        timedelta_obj = pd.to_timedelta(time_difference)
        days = timedelta_obj.days
        hours, remainder = divmod(timedelta_obj.seconds, 3600)
        minutes = remainder // 60
        print(f"Time difference: {days} days, {hours} hours, {minutes} minutes")

    except ValueError:
        print("Invalid date format. Please enter dates in the format YYYY-MM-DD.")

if __name__ == "__main__":
    main()
```

```python
#9
import pandas as pd

def main():
    try:
        
        file_path = input("Enter the path to the CSV file: ")

        
        df = pd.read_csv(file_path)

        
        column_name = input("Enter the name of the column to convert to categorical: ")

        
        category_order_str = input("Enter the category order (comma-separated): ")
        category_order = [cat.strip() for cat in category_order_str.split(",")]

        
        df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

        
        sorted_df = df.sort_values(by=column_name)

        
        print("\nSorted data:")
        print(sorted_df)

    except FileNotFoundError:
        print("File not found. Please check the file path.")
    except KeyError:
        print("Column not found in the CSV file.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
#10
import pandas as pd
import matplotlib.pyplot as plt

def main():
    try:
        
        file_path = input("Enter the path to the CSV file: ")


        df = pd.read_csv(file_path)

    
        sales_by_product = df.groupby(['Date', 'Product'])['Sales'].sum().unstack()

        
        sales_by_product.plot(kind='bar', stacked=True, figsize=(10, 6))
        plt.xlabel('Date')
        plt.ylabel('Sales')
        plt.title('Sales by Product Category Over Time')
        plt.legend(title='Product', loc='upper left')

        
        plt.show()

    except FileNotFoundError:
        print("File not found. Please check the file path.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()

```
